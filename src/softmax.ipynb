```python
import numpy as np
class Softmax:
    #Noting that keepdims means to keep the dimensions and maintain shape of the output aray vv
    def forward(self, x):
        #LSE trick
        x_stable = np.exp(x - np.max(x, axis = 1, keepdims=True)) #If x is a batch (in the shape of: [batch_size, num_classes]), then np.max(x) would only return a single scalar value, which is the same value subtracted from all rows
        exp_sum = np.sum(x_stable, axis=1, keepdims=True)
        self.y_pred = x_stable / exp_sum
        return self.y_pred

    # not assuming cross entropy
    def backward(self, d_out):
        elementSum = np.sum(d_out * self.y_pred, axis=1, keepdims=True)
        dx = self.y_pred * (d_out - elementSum)
        return dx

    
```
